```python
#import data
input_data= pd.read_csv("cancer_data.csv")
```

```python
X_variables=input_data.drop(['id','diagnosis'], axis=1).columns
```

```python
Y_variables='diagnosis'
```

```python
#Split train and test
X_train, X_test, Y_train, Y_test = train_test_split(input_data[X_variables],input_data[Y_variables], test_size=0.2,random_state=13)
```

```python
X_train
```

```python
#Grid Serach CV for SVM
def svc_param_selection(X, y, nfolds):
    param_grid = {'C': [0.001, 0.01, 0.1, 1, 10], 'gamma' : [0.001, 0.01, 0.1, 1]}
    grid_search = GridSearchCV(svm.SVC(kernel='linear'), param_grid, cv=nfolds)
    grid_search.fit(X, y)
    return grid_search.best_params_
```

```python
svc_param_selection(X_train,Y_train,5) #gamma is applicable for non-linear data.
```

```python
clf = svm.SVC(C=10,gamma=0.001,kernel='linear')
clf.fit(X_train,Y_train)
pred = clf.predict(X_test)
```

```python
metrics.accuracy_score(Y_test, pred)
```

```python
confusion_matrix(Y_test, pred)
```

```python
print(classification_report(Y_test, pred))
```
